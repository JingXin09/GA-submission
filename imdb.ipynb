#### Figure out something "interesting" using the actors data!

```python
#Each acor did how many movies?

#remove the unwanted word and symbol
def replace(string):
    return string.replace("[","").replace("]","").replace("u'","").replace("',",",")[:-1]
    
movies_series = movies['actors_list'].apply(replace)

#remove spaces in front of each actor name and split the name
actors_list = []
for movie_actors in movies_series:
    actors_list.append([i.strip() for i in movie_actors.split(',')])

#count the number of movie each actor act
actor_dict = {}
for actor in actors_list:
    for a in actor:
        if a in actor_dict:
            actor_dict[a] +=1
        else:
            actor_dict[a] = 1

actor_dict
```
